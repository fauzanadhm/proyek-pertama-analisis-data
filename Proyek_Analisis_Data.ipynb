```python
available_hour_every_day = hour_df.groupby(by="dteday").hr.count().reset_index()
```

```python
missing_hour = available_hour_every_day[available_hour_every_day["hr"] < 24]
```

```python
date_with_missing_hour = missing_hour["dteday"].values.tolist()
day_record_with_no_missing_hour = day_df[~day_df["dteday"].isin(date_with_missing_hour)]
day_record_with_no_missing_hour
```

```python
hour_record_with_no_missing_hour = hour_df[
    ~hour_df["dteday"].isin(date_with_missing_hour)
]
hour_record_with_no_missing_hour
```

**Insight:**

- day.csv berisi data yang lengkap, meng-cover sepanjang tahun 2011 dan 2012, tidak ada missing value
- hour.csv, meskipun tidak ada missing value, record tidak lengkap. Beberapa jam pada beberapa tanggal tidak ada datanya.
